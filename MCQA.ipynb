##__حذف لینک‌ها و ایموجی‌ها</center>__


<div dir='rtl'>
حذف لینک‌ها
</div>

```python
def removeLink(text):
  text = re.sub(r"http\S+", "", text)
  return text
```

<div dir='rtl'>
حذف کاراکتر‌های خاص و ایموجی‌ها
</div>

```python
emoji_pattern = re.compile("["
  u"\U0001F600-\U0001F64F"  # emoticons
  u"\U0001F300-\U0001F5FF"  # symbols & pictographs
  u"\U0001F680-\U0001F6FF"  # transport & map symbols
  u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
  u"\U00002500-\U00002BEF"  # chinese char
  u"\U00002702-\U000027B0"
  u"\U00002702-\U000027B0"
  u"\U000024C2-\U0001F251"
  u"\U0001f926-\U0001f937"
  u"\U00010000-\U0010ffff"
  u"\u2640-\u2642" 
  u"\u2600-\u2B55"
  u"\u200d"
  u"\u23cf"
  u"\u23e9"
  u"\u231a"
  u"\ufe0f"  # dingbats
  u"\u3030" # flags (iOS)
"]+", flags=re.UNICODE)

def removeEmojiesAndSymbols(text):
  text = emoji_pattern.sub(r'', text)
  text = re.sub('[!%#$,.`~!^&*()-+={}[]|\\//:;\"\'<,>?]؛۱۲۳۴۵۶۷۸۹۰1234567890', '', text)
  return text
```

```python
# pNormalizer = ParsivarNormalizer()
hNormalizer = HazmNormalizer()

def normalizeText(text):
  text = sub_alphabets(text)
  # text = pNormalizer.normalize(text)
  return hNormalizer.normalize(text)
```

```python
import math

def utils_preprocess_text(text):
  if not text or not isinstance(text, str):
    return ""
  text = removeLink(text)
  text = removeEmojiesAndSymbols(text)
  text = re.sub(r'[^\w\s]', '', str(text).strip())
  text = normalizeText(text)
  return text
```

#__اجرا</center>__


```python
queries = ["کدام گزینه در رابطه با آخرین ساخته اصغر فرهادی می‌باشد؟"]
docs = [
    "قهرمان فیلمی ایرانی به نویسندگی و کارگردانی اصغر فرهادی است که آخرین اثر این کارگردان است.",
     "آخر فیلم‌های اصغرفرهادی سبک خاصی از تولید فیلم است",
     "بهترین فیلم ساخته شده توسط اصغر فرهادی فیلم شعله بود",
     "آخرین جوایز جشنواره فجر به اصغر فرهادی داده شد"
]

queries = list(map(utils_preprocess_text, queries))
docs = list(map(utils_preprocess_text, docs))


doc_emb = model.encode(docs)

for query in queries:
  query_emb = model.encode(query)

  scores = util.dot_score(query_emb, doc_emb)[0].cpu().tolist()

  doc_score_pairs = list(zip(docs, scores))

  doc_score_pairs = sorted(doc_score_pairs, key=lambda x: x[1], reverse=True)

  print(query)
  for doc, score in doc_score_pairs:
      print(score, doc)
  print()
```
